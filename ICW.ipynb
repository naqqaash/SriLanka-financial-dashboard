```python
def plot_indicator(indicator_name):
    df_filtered = df[df["Indicator Name"] == indicator_name]
    plt.figure(figsize=(10, 5))
    plt.plot(df_filtered["Year"], df_filtered["Value"], marker='o')
    plt.title(indicator_name)
    plt.xlabel("Year")
    plt.ylabel("Value")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
```
